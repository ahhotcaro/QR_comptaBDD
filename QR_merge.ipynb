### Merge de deux dataframes

```python
import pandas
pandas.__version__
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
grand_livre = pd.read_excel("QR_rendu_2020-2021.xlsx")
```

```python
grand_livre
```

```python
plan_comptable = pd.read_excel("TR__Plan_comptable/Plan comptable.xls")
```

```python
plan_comptable
```

```python
grand_livre_V1 = grand_livre.merge(plan_comptable, how='left', on='plan_compta')
```

```python
grand_livre_V1
```

```python
grand_livre_V2 = grand_livre_V1[["date","plan_compta","Intitulé du compte général","debit","credit","solde progressif"]]
```

```python
grand_livre_V2
```

```python
# créer au préalable un fichier excel 
file_name = 'QR_rendu_2020-2021.xlsx'
```

```python
grand_livre_V2.to_excel(file_name)
print('Sales record successfully exported into Excel File')
```
