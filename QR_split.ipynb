### Split des plans comptables
#### permet de séparer les numéro des intitulés comptables

```python
grand_livre = pd.read_excel("QR_rendu_2020-2021.xlsx")
```

```python
grand_livre
```

```python
# supprimer le mot 'Suite' 
grand_livre['plan_compta'] = grand_livre['plan_compta'].str.replace('Suite ','')
```

```python
grand_livre['plan_compta'][48]
```

```python
# supprimer les chaines de caractères après le premier " " pour ne garder que le numéro compta
var_test = grand_livre['plan_compta'].str.split(' ').str[0]
```

```python
var_test
```

```python
grand_livre['plan_compta'] = var_test
```

```python
grand_livre
```

```python
# créer au préalable un fichier excel 
file_name = 'QR_rendu_2020-2021.xlsx'
```

```python
grand_livre.to_excel(file_name)
print('Sales record successfully exported into Excel File')
```
